## Question 1: Using the given MT cars dataset you need to find outlier with z score.

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
import os

os.chdir('E:\DataSets')
```

```python
df = pd.read_csv('mtcars.csv')
```

```python
df
```

```python
df.describe()
```

```python
data = list(df['hp'])
print(data)
```

```python
print(sns.boxplot(data))
```

```python
mean = np.mean(data)
```

```python
std_deviation = np.std(data)
```

```python
print('Mean :', mean)
print('Standard Deviation :', std_deviation)
```

```python
def outlier(data):
    thresholdValue = 2
    outlier = []
    for i in data:
        z_test = (i - mean) / std_deviation
        if z_test > thresholdValue:
            outlier.append(i)
    print(outlier)
```

```python
outlier(data)
```
